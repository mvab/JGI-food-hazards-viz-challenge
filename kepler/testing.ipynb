```python
from os import listdir
from keplergl import KeplerGl
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('../data/tidy/food_hazards_data_all.csv', sep='\t')
df.head()
```

```python
map1 = KeplerGl(height=600, width=800)
map1
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['long.origin'], df['lat.origin']))
map1.add_data(data=gdf, name="origin")
```
